### 리스트에 항목 추가하기

```python
# 1 ~ N의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = []
for i in range(1,N+1):
    num_list.append(i)
print(num_list)
```

```python
# 1 ~ N의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = [i for i in range(1,N+1)]
print(num_list)

# expression
num_list = [i * 2 for i in range(1,N+1)]
print(num_list)

# if문 사용
N = int(input("숫자 :"))
num_list = [i for i in range(1,N+1) if i % 2 == 0]
print(num_list)
```

### 리스트와 반복구조 연습

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def getIndex(num_list, target):
    for i in range(len(num_list)):
        if num_list[i] == target:
            return i
print(getIndex(number_list, 25))
```

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def getMax(num_list):
    max_val = num_list[0]
    for num in num_list:
        if num > max_val:
            max_val = num
    return max_val
print(getMax(number_list))
```

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def getMin(num_list):
    min_val = num_list[0]
    for num in num_list:
        if num < min_val:
            min_val = num
    return min_val
print(getMin(number_list))
```

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def countGT(num_list, target):
    count = 0
    for num in num_list:
        if num > target:
            count += 1
    return count
print(countGT(number_list, 42))
```

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def sumList(num_list):
    total = 0
    for num in num_list:
        total += num
    return total
print(sumList(number_list))
```

```python
number_list = [23, 45, 27, 11, 25, 65, 78]
def swapList(num_list):
    left = 0
    right = len(num_list) - 1

    while left < right:
        num_list[left], num_list[right] = num_list[right], num_list[left]
        left += 1
        right -= 1
print(number_list)
```
